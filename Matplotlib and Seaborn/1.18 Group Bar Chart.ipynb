```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_raw = pd.read_csv(
    "housing_data.csv",
    parse_dates=["period_begin", "period_end"],
    dtype={"total_homes_sold": "Int32", "region_name": "category"}
)

housing_raw.head()
```

```python
ca_or = housing_raw.loc[housing_raw["region_name"].str[-2:].isin(["CA", "OR", "WA"])]

ca_or.head()
```

```python
ca_or = ca_or.assign(
    price_range = pd.cut(
                    ca_or.loc[:,"median_active_list_price"],
                    [0, 400000, 600000, 2000000],
                    labels=["0-400k", "400k-600k","600k+"]),
    state = lambda x: x["region_name"].str[-2:]
        )

ca_or.head()
```

```python
import seaborn as sns
sns.barplot(
    x="price_range",
    y="inventory",
    hue="state",
    estimator=sum,
    data=ca_or
)
```

```python
ca_or = ca_or.query("period_end == '2021-08-01'").pivot_table(index="price_range",
             columns="state",
             values="inventory",
            aggfunc="sum"
         )

ca_or.head()
```

```python
ca_or.plot.bar()
```

```python
import numpy as np
fig, ax = plt.subplots()

width = .25

x1 = np.arange(len(ca_or))
x2 = [x + width for x in x1]
x3 = [x + width for x in x2]

ax.bar(
    x1, 
    ca_or["CA"], 
    label="California",
    width=width
)

ax.bar(
    x2, 
    ca_or["OR"], 
    label="Oregon", 
    width=width
)


ax.bar(
    x3, 
    ca_or["WA"], 
    label="Washington", 
    width=width
)

ax.set_title("Unit Sales of Homes in CA and OR")
ax.set_ylabel("Homes Sold")

ax.set_xlabel('Price Range')
# ax.set_xticks(x2)
# ax.set_xticklabels(ca_or.index)

plt.xticks(
    x2, 
    ca_or.index, 
)

ax.legend()

plt.show()
```
