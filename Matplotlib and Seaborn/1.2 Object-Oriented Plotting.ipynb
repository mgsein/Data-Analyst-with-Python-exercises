```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [x * 10 for x in range(10)]

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(data)
ax.set_title("Title")
fig.suptitle("Overall Title")
```

```python
fig, ax = plt.subplots()

ax.plot(data)
ax.set_title("Title")
fig.suptitle("Overall Title")
```
