```python
import pandas as pd
import numpy as np
```

```python
my_series = pd.Series(
        [0, 1, 2, 3, 4], index=['day0', 'day1', 'day2', 'day3', 'day4'] 
)

my_series
```

```python
my_series != 2
```

```python
my_series.loc[my_series > 2]
```
