```python
import pandas as pd

transactions = pd.read_csv('transactions.csv', parse_dates = ["date"])

transaction_44 = transactions.loc[transactions["store_nbr"] == 44, ["date", "transactions"]]
transaction_44.head()
```

```python
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")
```

```python
(transactions
.query("store_nbr in [44, 47] and date.dt.year == 2014")
.pivot(index="date", columns = "store_nbr")
.droplevel(0, axis = 1)
.plot(
    title = "Store 44 transaction 2013-2017",
    xlabel = "Date",
    ylabel = "Daily Transactions",
    color=["blue", "orange"],
    alpha=.3,
    style = ["-", "--"],
    subplots=True,
    layout=(1, 2),
    sharey=True,
    figsize=(10, 10),
     )
)
```

```python
transactions.groupby(["store_nbr"]).agg({"transactions": "sum"})
```

```python
transactions.groupby(["store_nbr"]).agg({"transactions": "sum"}).plot.bar(figsize=(10, 10))
```

```python
store_40s = list(range(40, 50))

(transactions
    .query('store_nbr in @store_40s')
    .groupby(["store_nbr"])
    .agg({"transactions": "sum"})
    .plot.bar(figsize=(10, 10))
)
```

```python
store_40s = list(range(40, 50))

(transactions
    .query('store_nbr in @store_40s')
    .groupby(["store_nbr"])
    .agg({"transactions": "sum"})
    .plot.barh(figsize=(10, 10))
)
```
