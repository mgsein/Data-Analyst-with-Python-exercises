```python
import pandas as pd
```

```python
transactions = pd.read_csv('transactions.csv', parse_dates = ["date"])

transactions.head()
```

```python
transactions.plot()
```

```python
transactions.set_index("date").loc["2013-01", "transactions"].plot()
```

```python
transactions.query("store_nbr == 44").set_index("date").loc["2013-01", "transactions"].plot()
```

```python
transaction_44 = transactions.loc[transactions["store_nbr"] == 44, ["date", "transactions"]]
transaction_44.head()
```

```python
transaction_44.set_index("date").plot()
```

```python
(transactions
.query("store_nbr in [44, 47]")
.pivot_table(index="date", columns = "store_nbr")
.droplevel(0, axis = 1)
.plot()
)
```
