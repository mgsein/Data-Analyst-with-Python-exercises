```python
import pandas as pd

transactions = pd.read_csv('transactions.csv', parse_dates = ["date"])

transaction_44 = transactions.loc[transactions["store_nbr"] == 44, ["date", "transactions"]]
transaction_44.head()
```

```python
transaction_44.set_index('date').plot(
    title = "Store 44 Transaction 2013-2017",
    xlabel = "Date",
    ylabel = "Daily Transactions",
    c="k"
)
```

```python
(transactions
.query("store_nbr in [44, 47]")
.pivot(index="date", columns = "store_nbr")
.droplevel(0, axis = 1)
.plot(color=["blue", "orange"],
#       cmap="viridis",
#       alpha=.4
     )
)
```
